```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to detect outliers based on percentiles
def percentile_based_outlier(gold_data, threshold=95):
    diff = (100 - threshold) / 2
    minval, maxval = np.percentile(gold_data, [diff, 100 - diff])
    return (gold_data < minval) | (gold_data > maxval)

# Column names for which we will detect outliers
col_names = ['SPX', 'GLD', 'USO', 'SLV', 'EUR/USD']

# Create a subplot with a specific size
fig, ax = plt.subplots(len(col_names), figsize=(12, 18))
fig.suptitle('Outlier Detection Using Percentile Method', fontsize=16, color='#004d40', fontweight='bold')

# Loop through each column and plot
for i, col_val in enumerate(col_names):
    x = gold_data[col_val][:1000]
    sns.kdeplot(x, ax=ax[i], shade=True, color='#00acc1')
    outliers = x[percentile_based_outlier(x)]
    ax[i].scatter(outliers, np.zeros_like(outliers), color='red', s=50, label='Outliers', clip_on=False)
    ax[i].set_title(f'Outlier detection - {col_val}', fontsize=14, color='#004d40')
    ax[i].set_xlabel(col_val, fontsize=12)
    ax[i].set_ylabel('Density', fontsize=12)
    ax[i].legend(loc='upper right')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score

def evaluate_model(model, X_train, X_test, Y_train, Y_test, model_name):
    model.fit(X_train, Y_train)
    predictions = model.predict(X_test)
    
    mae = mean_absolute_error(Y_test, predictions)
    mse = mean_squared_error(Y_test, predictions)
    rmse = np.sqrt(mse)
    r2_test = r2_score(Y_test, predictions)
    r2_train = r2_score(Y_train, model.predict(X_train))
    explained_variance = explained_variance_score(Y_test, predictions)
    
    return {
        "Model": model_name,
        "MAE (test)": mae,
        "MSE (test)": mse,
        "RMSE (test)": rmse,
        "R2 Score (test)": r2_test,
        "R2 Score (train)": r2_train,
        "Explained Variance (test)": explained_variance
    }

results = []

results.append(evaluate_model(regressor, X_train, X_test, Y_train, Y_test, "Random Forest"))
results.append(evaluate_model(decision_tree, X_train, X_test, Y_train, Y_test, "Decision Tree"))
results.append(evaluate_model(knn_reg, X_train, X_test, Y_train, Y_test, "K-Nearest Neighbors"))
results.append(evaluate_model(regressor_lr, X_train, X_test, Y_train, Y_test, "Linear Regression"))

results_df = pd.DataFrame(results)
results_df.style.background_gradient(cmap='Blues').set_properties(**{'font-family': 'Segoe UI'})
```
